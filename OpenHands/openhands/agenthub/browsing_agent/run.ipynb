```python
def search_in_web(goal: str) -> str:
    """
    Search the web for the goal and return the output.

    Args:
        goal (str): The goal to run

    Returns:
        str: The output of the goal
    """
    try:
        import requests

        API_KEY = "YOUR_API_KEY"
        CSE_ID = "YOUR_CSE_ID"
        query = goal
        url = "https://www.googleapis.com/customsearch/v1"
        params = {
            "key": API_KEY,
            "cx": CSE_ID,
            "q": query,
        }

        response = requests.get(url, params=params)
        response.raise_for_status()
        results = response.json()

        if "items" in results:
            output = "\n".join([item["snippet"] for item in results["items"]])
        else:
            output = "No results found."
        return output
    except Exception as e:
        return f"Error running goal: {e}"
```
